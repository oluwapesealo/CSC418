```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
    
names ={
    'Pese' : '0001',
    'Anderson' : '0002',
    'Lotanna' : '0006',
    'Chimaje' : '0003',
    'David' : '0004',
    'Kaobi' : '0005',
    'George' : '0007',
    'Lala' : '0008',
    'Zion' : '0009',
    'Desmond' : '0010'
    
}
def verify_username(username, matno):   
    
    if username in names:
        matno1 = names[username]
 
        if matno == matno1:
            image_technique(username)
        else:
            print(f"The Matriculation number inputed does not match username {username}")
    else:
        print(f"username {username} not found")
def image_technique(username):
    print("Select the enhancement you would like to use:\n ")
    it = input("For Normal, enter normal \nFor Sharpened, enter sharpened \nFor Brighteness and contrast, enter bandc \nFor Median Blur, enter mandb \nFor Scaled, enter scaled \nFor Inverse, enter inverse \nFor Weighted, enter weighted\n")
    if it == "normal":
        normal(username)
    if it == "sharpened":
        sharpening(username)
    if it == "bandc":
        brightness_contrast(username)
    if it == "mandb":
        median_blur(username)
    if it == "scaled":
        scaled(username)
    if it == "inverse":
        inverse(username)
    if it == "weighted":
        weighted(username)
    else:
        print("Invalid input, Please try again")
 #   input("")
 #   input("")
 #   input("")
 #   input("")
 #   input("")
#   input("")
    
def normal(username):
    path = f'images/portraits/{username}.jpg'
    image = cv2.imread(path)

    cv2.imshow("original", image)
    cv2.waitKey(0)

    cv2.destroyAllWindows()
    
def brightness_contrast(username):
    path = f'images/portraits/{username}.jpg'
    image = cv2.imread(path)

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)

    brightness = 5
    contrast = 1.5
    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)

    cv2.imwrite(f'images/portraits/{username}brightandcontrasted.jpg', image2)

    plt.subplot(1, 2, 2)
    plt.title("Brightness & contrast")
    plt.imshow(image2)
    plt.show

    
    
def sharpening(username):
    path = f'images/portraits/{username}.jpg'
    image = cv2.imread(path)
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)

    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

    sharpened_image = cv2.filter2D(image, -1, kernel)

    cv2.imwrite(f"images/portraits/{username}sharpened_image.jpg", sharpened_image)

    plt.subplot(1, 2, 2)
    plt.title("Sharpening")
    plt.imshow(sharpened_image)
    plt.show
    
def median_blur(username):
    path = f'images/portraits/{username}.jpg'
    image = cv2.imread(path)

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)

    filtered_image = cv2.medianBlur(image, 15)

    cv2.imwrite(f'images/portraits/{username}Median-Blur.jpg', filtered_image)

    plt.subplot(1, 2, 2)
    plt.title("Median Blur")
    plt.imshow(filtered_image)
    plt.show

def scaled(username):
    path = f'images/portraits/{username}.jpg'
    image = cv2.imread(path)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)

    scaled_image = cv2.resize(image, None, fx = 2, fy =2)
    cv2.imwrite(f'images/portraits/{username}scaled.jpg', scaled_image)

    plt.subplot(1, 2, 2)
    plt.title("Scaled")
    plt.imshow(scaled_image)
    plt.show
    
def inverse(username):
    path = f'images/portraits/{username}.jpg'
    image = cv2.imread(path)

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)

    inverse_image = 255 - image

    cv2.imwrite(f'images/portraits/{username}inverse_image.jpg', inverse_image)

    plt.subplot(1, 2, 2)
    plt.title("Inverse color")
    plt.imshow(inverse_image)
    plt.show

def weighted(username):
    username2 = input("please enter another username to add: \n")
    path1 = f'images/portraits/{username}.jpg'
    path2 = f'images/portraits/{username2}.jpg'
    image10 = cv2.imread(path1)
    image20 = cv2.imread(path2)
    

    image1 = cv2.resize(image10, (500,400))
    image2 = cv2.resize(image20, (500,400))

    addImage = cv2.addWeighted(image1, 0.5, image2, 0.6, 0)


    cv2.imshow('Weighted Image', addImage)

    if cv2.waitKey(0) & 0xff == 27:
        cv2.destroyAllWindows()

def program():
    username = input("Please enter your username [Your First name]: ")
    matno = input("Please enter your Matriculation Number: ")
    verify_username(username, matno)
    
    
program()
```
