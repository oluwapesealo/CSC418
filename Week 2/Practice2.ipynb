```python
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('images/sst.jpg')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

brightness = 5
contrast = 1.5
image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)

cv2.imwrite('images/contrast_image.jpg', image2)

plt.subplot(1, 2, 2)
plt.title("Brightness & contrast")
plt.imshow(image2)
plt.show
            
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('images/sst-foyer.jpg')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

sharpened_image = cv2.filter2D(image, -1, kernel)

cv2.imwrite("images/sharpened_image.jpg", sharpened_image)

plt.subplot(1, 2, 2)
plt.title("Sharpening")
plt.imshow(sharpened_image)
plt.show
            
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('images/sst-foyer.jpg')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

filtered_image = cv2.medianBlur(image, 15)

cv2.imwrite('images/Median-Blur.jpg', filtered_image)

plt.subplot(1, 2, 2)
plt.title("Median Blur")
plt.imshow(filtered_image)
plt.show

```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('images/sst.jpg')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

scaled_image = cv2.resize(image, None, fx = 2, fy =2)
cv2.imwrite('images/scaled.jpg', scaled_image)

plt.subplot(1, 2, 2)
plt.title("Scaled")
plt.imshow(scaled_image)
plt.show
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('images/sst-foyer.jpg')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

inverse_image = 255 - image

cv2.imwrite('images/inverse_image.jpg', image)

plt.subplot(1, 2, 2)
plt.title("Inverse color")
plt.imshow(inverse_image)
plt.show
```
